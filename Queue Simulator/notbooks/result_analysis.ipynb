```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
def plot_metrics(path="../data/metrics/all_metrics.csv"):
    sns.set(style="whitegrid")

    df = pd.read_csv(path)
    total_latency = df['wait_time'] + df['generation_time']

    # --- Figure 1: Queue & Latency Over Time ---
    fig, ax = plt.subplots(2, 1, figsize=(12,8), sharex=True)

    # Queue length
    ax[0].plot(df.index, df['queue_length'], marker='o', linestyle='-', color='blue')
    ax[0].set_ylabel("Queue Length")
    ax[0].set_title("Queue Length Over Time")

    # Total latency
    ax[1].plot(df.index, total_latency, marker='o', linestyle='-', color='red')
    ax[1].set_xlabel("Request Index")
    ax[1].set_ylabel("Total Latency (s)")
    ax[1].set_title("Total Latency Over Time")

    plt.tight_layout()
    plt.show()

    # --- Figure 2: Distribution of Metrics ---
    fig, ax = plt.subplots(1, 3, figsize=(18,5))

    # Waiting time
    sns.histplot(df['wait_time'], bins=20, kde=True, color='green', ax=ax[0])
    ax[0].set_xlabel("Wait Time (s)")
    ax[0].set_title("Waiting Time Distribution")

    # Generation time
    sns.histplot(df['generation_time'], bins=20, kde=True, color='orange', ax=ax[1])
    ax[1].set_xlabel("Generation Time (s)")
    ax[1].set_title("Generation Time Distribution")

    # CLIP self-similarity
    sns.histplot(df['clip_self_similarity'], bins=20, kde=True, color='purple', ax=ax[2])
    ax[2].set_xlabel("CLIP Self-Similarity")
    ax[2].set_title("Image Quality Distribution")

    plt.tight_layout()
    plt.show()
    df.describe().to_excel("describe_output.xlsx")
```

```python
plot_metrics(path="../data/metrics_SDXL-turbo (Stable Diffusion XL turbo) 1.1.1/all_metrics.csv")
```
